#### Checking For Outliers



```python
# List of columns for which we want to create histograms
columns_to_plot = ["session duration", "avg order value", "promotion clicks", 
                   "sale product views", "app transactions", "account length", 
                   "add to wishlist", "desktop sessions", "customer service calls"]
columns_to_plot.sort()

# Define a list of colors for the histograms
colors = ['navy', 'darkorange', 'limegreen', 'tomato', 'dodgerblue', 'coral', 'mediumseagreen', 'lightcoral', 'cornflowerblue']

# Create a figure with subplots
fig, axes = plt.subplots(3, 3, figsize=(20, 20))

# Plot histograms for all nine columns
for i, column in enumerate(columns_to_plot):
    row, col = divmod(i, 3)
    axes[row, col].hist(c_data[column], bins=15, color=colors[i])
    axes[row, col].set_title(f'Histogram of {column}')
    axes[row, col].set_xlabel(column)
    axes[row, col].set_ylabel("Frequency")
    axes[row, col].grid(True, linestyle='--', alpha=0.6)
```

```python
c_data.describe()
```

```python
c_data_new = c_data[["session duration", "avg order value", "promotion clicks", 
                   "sale product views", "app transactions", "account length", 
                   "add to wishlist", "desktop sessions", "customer service calls"]]

quartiles = c_data_new.quantile([0.25, 0.75])
IQR = quartiles.loc[0.75] - quartiles.loc[0.25]

IQR_sorted = IQR.sort_index()

Q1 = quartiles.loc[0.25]
Q3 = quartiles.loc[0.75]

lower_limit = Q1 - 1.5 * IQR_sorted
upper_limit = Q3 + 1.5 * IQR_sorted

# The IQR for each column is now stored in the 'IQR' Series
print(f'IQR values for columns:\n{IQR_sorted} \n')
print(f'lower_limit for columns:\n{lower_limit}\n')
print(f'upper_limit for columns:\n{upper_limit}')
```

```python
Number_of_outliers = ((c_data_new < lower_limit) | (c_data_new > upper_limit)).sum()
print("Number of outliers for each column:")
print(Number_of_outliers)
```

# 8. Analysis of the Dependent variable (Churn) vs Idependent variables

Until now we've looked into the numeric variables, their missing data, duplicates, distribution, and outliers. A key part of the customer behaviour data is to understand which "behaviours" aka idependent varialbes in our dataset can help us predict customers deciding to stop using the product, aka "churn".

Another variable that we haven't analysed yet is the location code, but we might be able to find value in comparing it with the churn and finding out any potential correlations.

### Pie Chart to visualize the overall Churn rate

```python
churn_counts = c_data['churn'].value_counts()
labels = ['Retained Customers', 'Churned Customers']
explode = (0.1, 0.2) 
wp = { 'linewidth' : 1.5, 'edgecolor' : "black" }

# Create a pie chart with additional features
fig, ax = plt.subplots()
ax.pie(churn_counts, labels=labels, autopct='%.02f%%', explode=explode, colors=colors, startangle=45, wedgeprops = wp)

# Display the pie chart
plt.show()
```

# 9. Plot Churn against the categorical data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a countplot for location code vs Churn
plot = sns.countplot(data=c_data, x='location code', hue='churn', palette='Set2')

# Add counts above the bars
for p in plot.patches:
    plot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                  textcoords='offset points')

# Set plot title and labels
plt.title('Location Code vs Churn')
plt.ylabel('Number of Customers')
plt.xlabel('Location Code')

# Customize legend labels
plot.legend(title='Churn', labels=['Retained Customers', 'Churned Customers'])

plt.show()
```

From this, we can say that location code 415 has majority of customers

```python
# Countplot for Push Notifications status vs Churn
plot = sns.countplot(data=c_data, x='push status', hue='churn')

# Added counts above the bars
for p in plot.patches:
    plot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center', fontsize=14, color='black', xytext=(0, 5),
                  textcoords='offset points')

# Set plot title 
plt.title('Push Notifications Status')
plt.ylabel('Number of Customers')
plt.xlabel('Push Status')

# Customize legend labels
plot.legend(labels=['Retained Customers', 'Churned Customers'])

plt.show()
```

From this, we can say that the majority of customers do not allow push notifications

```python
# Countplot for credit card info save Status vs Churn
plot = sns.countplot(data=c_data, x='credit card info save', hue='churn', palette='Set1')

# Added counts above the bars
for p in plot.patches:
    plot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center', fontsize=14, color='black', xytext=(0, 5),
                  textcoords='offset points')

# Set plot title 
plt.title('credit card info save vs Churn')
plt.ylabel('Number of Customers')
plt.xlabel('credit card info save status')

# Customize legend labels
plot.legend(labels=['Retained Customers', 'Churned Customers'])

plt.show()
```

From this, we can say that the majority of customers dont store their credit card information on the side.

# 9.3 Observations from plotting categorical data against Churn

- The location codes have a similar percentage of churns, meaning that location might not be a strong indication for churn
- Customers with push status activated are less likely to be churns, meaning that motivating customers to activate push status might help minimize churns
- Customers with their credit card information saved on the side are far less likely to be churns than customers without their credit card info saved. This means that if the shop motivates people to save their credit card info might minimize churns. 
    - This makes sense since customers who trust their credit card info on a side indicate that they trust the brand. More trust less chances of customers leaving your business.
